```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from math import floor
from pathlib import Path
import json

__rows = 1
__cols = 1


def sb_init(rows, cols):
    plt.figure(figsize=[cols*5, rows*4])
    global __rows
    global __cols
    __rows = rows
    __cols = cols
    return plt.subplot(rows, cols, 1)


def sb(x, y):
    return plt.subplot(__rows, __cols, x + (y - 1) * __cols)
```

```python
# Load data

p = Path('./')
dirs = [f for f in p.iterdir() if f.is_dir()]
data = []

for d in dirs:
    item = {}
    temp = np.loadtxt(d / 'CF_H_avg.txt')
    item['x'] = temp[:, 0]
    item['avg_h'] = temp[:, 1]
    details = (d / 'details.txt').read_text()
    if details[0] == 'S':
        # Old version without JSON formatting
        details = details.replace('\n', ', ') \
                         .replace("Side", "{\"side\"") \
                         .replace("Defects frac", "\"defects_frac\"") \
                         .replace("Gamma", "\"gamma\"") \
                         .replace("Replies", "\"replies\"") \
                         .replace("Corr. Range", "\"corr_range\"")
        details = details[:-2] + " }"
    item.update(json.loads(details))
    data.append(item)
```

```python
lims = [1, 30]
defects_frac = 0.5

xp = np.sort( data[0]["x"][1:] )

def expected( x, gamma ):
    return np.power( 1 + x * x, - gamma / 2 )

for i in data:
    if i["defects_frac"] == defects_frac:
        a1 = plt.loglog( i["x"], i["avg_h"], '.', label= r"$\gamma$ = " + str(i["gamma"] ))
        
        plt.loglog( xp, expected( xp, i["gamma"]), '-', color = a1[0].get_color(), label= "_hidden" )

plt.legend();

plt.title("Correlation function - h field");
plt.xlabel(r"$r$");
plt.ylabel(r"$<h(x)h(x+r)>$");

plt.savefig( "CF_h.pdf" )
plt.savefig( "CF_h.png" )
```
