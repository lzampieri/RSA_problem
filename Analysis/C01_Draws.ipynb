```python
import numpy as np
from functions import DrawsLoader, splt
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from math import sqrt

savedir = "C01/"
```

```python
data = DrawsLoader.load_data( "AcerAnalysis/default_20220515_24" )
if( len( data ) > 0 ):
    data = data[ np.vectorize( lambda i: i['gamma'] )(data).argsort() ]
else:
    print("No data!")

colors = {
    0: 'tab:blue',
    1: 'green',
    -1: 'tab:red'
}
get_cols = np.vectorize( lambda x: colors[x] )
```

```python
# Show data

splt.init( 1, len( data ) )
for d in data:
    splt.next()
    plt.scatter( d['draw']['x'], d['draw']['y'], marker='s',  color= get_cols( d['draw']['v'] ) )
    plt.axis( 'square' )
    plt.axis('off')
```

```python
# Save data

for d in data:
    plt.figure( figsize = (5, 5) )
    plt.scatter( d['draw']['x'], d['draw']['y'], marker='s',  color= get_cols( d['draw']['v'] ) )
    plt.axis( 'square' )
    plt.axis('off')
    plt.tight_layout()
    plt.savefig( savedir + "defs_" + str( d['gamma'] ) + ".pdf" )
```
